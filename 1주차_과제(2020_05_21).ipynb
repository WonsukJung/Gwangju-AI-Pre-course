# **인공지능을 사용하는 4가지 분야 적용사례**


###1. 서론

전세계적으로 인공지능(Artificial Intelligence/AI)는 현재 무수히 많은 제품과 서비스에 적용되어 있다. 특히 국내에서는 2016년 서울에서 개최된 바둑대결이 큰 화제가 되었는데, 이 대국은 구글(Google)에서 개발한 인공지능 바둑프로그램 알파고(AlphaGo)와 우리나라를 대표하는 바둑기사 이세돌과의 대결이었다. 이 계기를 통해 인공지능에 대한 한국인들의 관심이 급증하였고, 미래사회에서 인공지능이 우리 실생활에 어떠한 영향을 끼칠지에 대한 다양한 분석이 언론 보도와 수많은 관련 기관의 연구를 통해  활발히 이루어졌다. 인공지능 기술이 가장 보편적으로 적용된 4가지 분야의 사례를 통해 각 분야별로 주로 사용된 기술은 무엇이며 어떤 단계까지 발전했는지 살펴보고자 한다.

###2. 인공지능기술 기반 각 분야별 적용 사례

__1)	언어__  "챗봇(Chatbot) / RochaBot"

언어분야에서는 자연언어처리기술(NLP)을 기반으로 한 딥러닝 기술 발전에 따라 인공지능 서비스인 대화형 챗봇이 전 산업으로 확대되어 개발, 보급되고 있다. 대화형 챗봇은 경제, 사회생활, 교육, 산업, 기업 마케팅, 관공서 민원업무, 금융업, 보험업 등 산업 전반에 걸쳐 매우 유용한 기술로 각광받고 있다. 대화형 챗봇은 일반적인 로봇과는 달리 하드웨어적인 요소가 필요하지 않고, 소프트웨어적인 요소들로 제작이 가능하기 때문에 제작비가 저렴하다. 

자연어 처리의 근간을 이루는 언어 이해 기술은 ‘형태소 분석’, ‘구문 분석’, ‘개체명 분석’, ‘화행 분석’, ‘의도 분석’의 과정으로 이루어진다. 주로 규칙에 기반을 둔 방법과 기계 학습을 이용한 방법을 사용하는데, 최근에는 최신 기계 학습 방법으로 좋은 성능을 보이고 있는 딥러닝을 많이 사용하게 된 것이다. 예를 들면, 대화형 인공지능 챗봇인 RochaBot은 ㈜페르소나시스템이 2017년에 원천기술을 개발한 것으로, 금융, 키오스크, 보험 판매업 등에 사용되고 있다. 이 챗봇은 24시간 텍스트나 음성으로 인간과 대화자 AI가 대화를 통해 필요한 의사를 주고받을 수 있는 도구로서, 사용자가 필요로 하는 서비스와 데이터를 적시에 찾아주며 기업, 서비스, 개인들 간을 연결해주는 역할을 하고 있어서 일상생활과 비즈니스에 유용한 인공지능형 도구로 자리 매김을 했다. 

RochaBot 엔진은 **NLU(Natural Language Understanding)**와 **NLG(Natural Language Generation)**으로 구성되어 있으며, 이 두 종류의 자연언어처리기술을 사용하여 **DNN/RNN(신경망 체계)**과 같은 딥러닝 방식으로 자연어를 인식하는 기술을 사용하고 있다. RochaBot시스템은 데이터 소스, 빅데이터 플랫폼, AI 플랫폼의 3가지 부문으로 구성되어 있다. 데이터 소스는 챗봇에 사용되는 다양한 DB로서 이 데이터는 빅데이터 플랫폼 부문으로 연동된다. 빅데이터 플랫폼은 데이터 수집, 데이터 처리, 데이터 분석기로 구성되어 AI 플랫폼 작동을 직•간접으로 지원한다. AI 플랫폼은 온톨로지와 서비스로 구분되는데, 온톨로지 영역 내에서는 사용자 의도파악 셋트, 구체적인 정보요구 파악 셋트, 상황파악용 셋트, 대화 셋트을 내장하고 있다. 

RochaBot 엔진은 딥러닝기반(DNN/RNN)과 통계기반으로 구동되는데, 대화이해장치(NLU)에서 자연어를 인식하게 하고, 대화 자동생성장치(NLG)는 사전에 데이터를 학습하여 풍부한 답변 셋트를 구성할 수 있으며, 내장된 대화 엔진(Dialogue Engine)에서는 챗봇의 대화시나리오를 제작하여 적절한 응답문(정보)를 사용자 디바이스로 내보내도록 되어 있다.

대화형 챗봇을 이용하여 사용자와 챗봇 간의 대화가 원활하게 진행되기 위해서는 목적지향적인 대화 알고리즘의 고안이 필요하고 현재 다양한 종류의 대화형 챗봇이 개발되고 있지만 그 개발 기술의 고도화 수준에 따라서 챗봇 성능이 다르기 때문에 기술개발 수준을 고찰해 보는 것이 바람직하다.

__2)	음성__ – "음성인식 AI / 구글 홈, 아마존 에코 등"

음성분야는 최근 많은 기업에서 다양한 제품들을 경쟁적으로 출시하고 있어 소비자들에게 매우 익숙한 인공지능 기술 분야이다. 음성인식 기술은 그 발전과정을 5개의 단계로 구분한다고 하는데, 이 과제에서는 자세한 내용은 생략하고 2011년 이후 5세대 단계만을 중점적으로 설명하고자 한다. 5세대에서는 실시간으로 딥러닝 기술을 활용하여 사용자의 감정 등을 파악하는 수준의 연구가 진행되고 있어 구글 홈, 아마존 에코, SK 누구, KT 기가 지니, 애플 시리, 삼성 빅스비 등과 같은 제품이나 서비스들도 이 세대의 산물이라 볼 수 있다.

음성인식 관련 기술은 크게 두 가지이다. 음성을 텍스트로 변환하는 기술**(Speech to Text, 이하 STT)**과 텍스트를 음성으로 변환하는 기술**(Text to Speech, 이하 TTS)**이다. 또한 IBM, 구글, 아마존, 마이크로소프트 등 4종의 클라우드 플랫폼들은 다국어 인식이 가능하다. 영어, 독일어, 프랑스어, 스페인어, 일본어 등의 주요 언어를 실시간으로 인식하고 음성이나 텍스트로 변환한다. 구글은 110종 이상의 언어를 제공한다. TTS의 경우 언어별로 남성과 여성의 음성을 구분하여 제공하는 경우도 있다. TTS 서비스들은 단어 하나하나의 발음과 어조, 속도, 음량, 샘플레이트 조정 등 사람처럼 말하는 것을 목표로 하고 있다. 구글은 부적절 콘텐츠를 실시간으로 인식하여 필터링하거나 주변의 소음 속에서도 음성을 인식하는 기술을 제공한다. STT와 TTS의 결과는 대부분 변환된 텍스트와 변환된 음성이다. 아마존 웹 서비스의 경우 음성으로부터 변환된 텍스트에 주석을 달아 음성 응용프로그램이나 개발자가 마크업을 통해 발음, 음량, 피치, 속도 등을 지정할 수 있도록 하고 있다. 이 기술은 W3C 표준인 SSML(Speech Synthesis Markup Language) 형태의 마크업 언어이다. 

음성인식 성능은 음성 데이터베이스의 크기와 품질에 비례하여 성능향상이 이루어진다. 상용 서비스에 적용되는 음향 모델은 대부분 확률 통계 방식은 HMM(Hidden Markov Model) 기반으로 이루어졌으며, 2010년대에 들어서면서 딥러닝 기반의 HMM/DNN방식으로 단어 인식 오류율 기준으로 약 20% 정도의 성능 향상을 이끌어냈다. 최근에는 시퀀스-투-시퀀스(sequence-to-sequence)방식의 **RNN(Recurrent Neural Network)** 기반으로 속도와 성능 면에서 좋은 결과를 내기 시작했다. 

음성인식 AI기술은 우리 생활과 매우 밀접한 분야들에 유용하게 사용될 수 있다. STT 기술을 이용하여 회의 녹음파일을 텍스트로 추출하게 되면 회의록을 스크립트로 작성하는 공무원들의 기존 업무를 점차 경감시킬 것이고, 기록물관리 기관의 경우에는 이미 소장하고 있는 녹취되지 않은 기록물의 일괄 텍스트 추출을 통해 내용 기반의 검색 서비스 등으로 확장시킬 수 있다.  TTS 기술의 경우 기관 홈페이지 내 시각장애인을 위한 음성 서비스에 적용할 수 있다. 모바일 어플리케이션에 탑재할 경우 활용도는 더욱 증가할 것이다.

__3)	이미지__  - 구글 렌즈

인공지능(AI) 이미지 인식 기술은 지난 2012년 구글의 딥러닝 기반 고양이 분류 시험인 ‘Brain Project'를 시작으로 2014년 페이스북의 사람 얼굴 인식 기술인 '딥페이스(DeepFace)', 2015년 바이두의 이미지 인식 기술인 '딥이미지(Deep Image)' 등 IBM, MS, 구글 등 글로벌 IT기업들의 경쟁적인 개발과 잇단 플랫폼 출시를 거듭하는 동안 급격히 진화돼 인공지능 이미지 인식 기술은 산업, 기업, 경제, 사회 등 다양한 분야에서 우리의 삶 속으로 들어오고 있다.

이미지 분야의 가장 기본이 되는 딥러닝 기술은 전통적인 **CNN(Convolutional Neural Network)** 및 CNN의 단점을 개선한 **RNN(Recurrent Neural Network)**이다. 일반적으로 CNN은 고차원 데이터의 특징(feature)을 추출하는 convolution layer와 pooling layer, 최종적으로 데이터를 분류하는 fully-connected layer로 구성되며 convolution layer와 pooling layer는 사용자가 풀고자 하는 문제에 따라 임의로 각 layer의 개수와 순서를 조정할 수 있다는 장점이 있다. 즉 특화된 필터를 사용해 이미지 중 특정 색상이나 톤으로 이루어진 특징을 추출하고, 이 필터와 입력 데이터 상의 합성곱을 해주면 각 필터의 특성에 맞게 강조된 이미지의 특징을 추출할 수 있다. 그리고 이러한 작업을 여러 번 거치면서 최종적으로 CNN은 이미지를 인식하게 된다. CNN은 구조적인 특징으로 컴퓨터 비전 분야에서 특히 많이 활용되고 있다. 이미지 인식과 같은 기초적인 작업을 포함해, 이미지 복원, 이미지 설명, 움직임 추정 등 이미지를 활용한 보다 고차원적인 분석 문제에서 다른 알고리즘에 비해 독보적인 성능을 보이고 있다.

Recurrent Neural Network(RNN)는 일반적인 인공 신경망과 달리 순서가 있는 정보를 입력 데이터로 활용하는 신경망을 의미한다. 즉, 일반적인 신경망의 경우 입출력 정보가 서로 독립적이라고 가정하고 있는데, 이는 실제로 많은 현실 문제를 정확하게 모델링하는데 뚜렷한 한계를 가진다. RNN은 일반적인 인공 신경망과 달리 recurrent라는 용어가 그 의미를 내포하듯, 이전에 입력된 데이터와 지금 입력된 데이터를 동시에 고려해 출력 값을 결정한다. 즉, 각 층마다 다른 가중치를 가지는 일반적인 인공 신경망과는 달리, RNN은 시간의 흐름에 따라 모두 같은 가중치를 공유함으로써 이전 상태의 입력 값을 고려한 학습이 가능하다는 특징이 있다. 특수한 구조로 인해 RNN은 많은 자연어 처리 문제에 성공적으로 적용되어 왔다. 대표적으로 음성 인식이나 문장 생성, 기계 번역 등의 분야에서 효과적으로 활용되고 있으며, CNN 등과 결합해 이미지의 캡션이나 자막 생성에도 적용되고 있다.

구글에서 개발한 구글 렌즈는 보여지는 이미지를 읽고 더 나아가 정보를 이해하며 상호작용까지 가능한 똑똑한 카메라 어플리케이션이다. 이미지에 보여지는 정보를 분석하고 사용자에게 추가적인 정보를 빠르게 보여주는 것은 물론 확장된 기능에 실시간으로 연결할 수 있다. 구글 렌즈가 특별한 이유는 단순히 정보를 읽거나 이해함에서 그치지 않고 한 단계 더 나아가 이미지에서 습득한 정보를 분석하고 활용하여 스스로 필요한 추가적인 정보를 찾는다는 점이다. 즉, 이미지나 동영상에서 얻은 정보를 바탕으로 사용자의 의도를 파악하고 그에 따라 어떤 행동을 스스로 취할 수 있다는 것이다.

그러나 이미지 인식을 활용하는 구글 렌즈가 직면한 도전과제는 음성인식 기술이 가진 문제점과 비교해 더 복잡하다. 구글 음성인식 기술이 인식해야 하는 영어 단어 개수가 18만개 정도인 반면 구글 렌즈가 인식해야 하는 이미지 인식 대상이나 사물은 수십억가지에 달하며 각도, 광원 등에 따라 또 다르게 보일 수 있기 때문에 실제 인식해야 하는 객체 수는 더 방대하다. 구글은 분류, 탐지, 임베딩(embedding), 모델 학습 등의 기술적 접근으로 구글 렌즈의 정확도를 높여나가고 있다.

__4)	자율주행__ - 테슬라 모델3

4차 산업현명의 핵심 키워드는 ‘융합’이라고 흔히 이야기 한다. 현재 가장 주목 받고 있는 산업의 융합 분야는 아마 자동차와 인공지능(AI)가 만난 자율주행 자동차라고 할 수 있다. 현재 전 세계 자동차 회사들은 모두 경쟁적으로 인공지능 자율주행 자동차 개발에 뛰어 들어 제품개발에 열을 올리고 있다. 우리나라에서도 사람의 신경망을 모사한 딥러닝 기술을 이용하여 자율주행 자동차에 탑재할 수 있는 기술이 개발되었다.

자율주행 기능은 크게 인지, 판단 및 제어로 구성된다고 할 수 있다. 인지는 주변 차량, 행인, 차선, 교통신호, 장애물 등의 주변 상황, 위치, 속도, 바퀴 등의 차량 상태 같은 주행에 필요한 데이터를 수집하는 단계이며 카메라, 레이더, 라이다(LIDAR) 등 센서, GPS, 차량 통신(V2X: Vehicle to Everything) 등의 기술이 요구된다. 판단은 차량에서 수집된 데이터와 지도, 교통량, 도로 상황 등 환경 데이터를 종합하여 최적의 주행경로와 속도를 결정하는 단계로, 관련 기술은 인공지능 알고리즘, 빅데이터 분석, 전용 프로세서(GPU), 고정밀/3D 지도 등이다. 마지막으로 제어는 엔진, 브레이크 등 자동차의 각 부분을 제어하여 실제로 움직임을 구현하는 단계로 조향, 가감속, 제동 등 기존 자동차 기술이 활용된다.

자율주행 기술은 라이다(LIDAR), 레이다, 카메라 등의 센서로부터 얻은 데이터를 가공하여 차량을 제어하는 기술이 핵심이다. 특히, 카메라 센서는 전방 사물 인식 및 검출, 차선 인식 및 유지 기능, 신호등, 표지판, 도로 검출 등의 복합 환경 인식 등 사람의 눈과 같은 역할을 수행한다. 이러한 영상 인식기는 예전에는 사람이 직접 특징을 정의하여 영상으로부터 해당 특징을 추출하여 인식을 수행하는 방식이었으나, 최근에는 사람의 뇌의 신경망을 모사한 딥러닝 기술을 이용하여 스스로 의미 있는 특징을 추출하고 이를 이용하여 인식을 수행하는 방식이 널리 사용되고 있다. **딥뉴럴 네트워크(DNN, Deep Neural Network)**은 인간의 신경망을 모방한 네트워크로서 이전의 출력 값이 다음 층의 입력으로 작동되는 계층 구조를 갖는다. 매 계층의 출력단에는 비선형 함수가 존재하며, 모든 연결에는 가중치가 있다. 이 가중치는 학습에 의해 구해지게 되는데, 이 과정은 최상위 계층의 출력 값으로부터 오류를 계산하고 하위 계층으로 오류를 전파하는 방식이다. 이러한 DNN을 학습시키고 이용하는 기술을 딥러닝 기술이라고 한다. DNN이 구해야 하는 가중치의 수는 무수히 많기 때문에, 같은 계층에서는 가중치를 공유하는 방식도 개발되었다. 이를 **컨볼류션 뉴럴 네트워크 (CNN: convolutional neural network)** 라고 불리는데, 이 방식은 영상 인식에 가장 많이 쓰이는 기법이다. 이러한 CNN 기법은 더 복잡한 문제를 해결하기 위해서는 100층 이상 쌓기도 하며, 계산량이 상당하기 때문에 수천개의 계산을 동시에 할 수 있는 GPU를 이용하곤 한다. 차량에서는 NVIDIA가 개발한 drive PX를 활용하여 기존에는 계산량이 많아 처리하기 힘든 계산을 차량 환경 에서도 계산이 가능해졌다. 따라서, 최근 딥러닝 기반 인식 기술을 차량에 적용하는 사례가 많아지게 되었다.

자율주행을 위한 가장 기본적인 기능 중의 하나는 주변 상황의 물체를 탐지하는 기술이다. 전방에 있는 차량 혹은 보행자를 검출하고 자동으로 차량의 제동 시스템을 제어하여 사고를 미연에 방지할 수 있는 긴급 제동 시스템(AEB: autonomous emergency braking system) 등에 활용할 수 있는 기술이다. 기존에는 개발자들이 직접 만든 특징을 이용하여 분류기를 학습하고, 학습된 분류기를 이용하여 주변 환경에 있는 물체를 검출하고 인식하였으나, 성능이 만족스럽지 않아 한동안 주춤했었다. 하지만, 앞서 이미지 분야에서 설명했듯이 CNN을 이용하여 물체 검출하고 인식을 수행하는 방식이 사람만큼이나 높은 인식률을 보이고 있어 이를 자율 차량에도 적용하고 있는 추세이다. 딥러닝을 이용한 물체 검출 기법은 최근에도 연구가 진행 중에 있으며, 그 중 가장 혁신적인 기술로 손 꼽히는 기법은 **R-CNN(Regions with CNN features)**이라는 기법이다. 일반적으로 물체 검출에 CNN 을 적용하려면 영상의 모든 영역 및 모든 크기에 대해 탐색을 수행하여야 한다. 하지만 이 기법을 이용하면 처리속도를 개선시킬 수 있는데, 기본적으로 영역제안(region proposal)을 2,000개 정도 추출하여 추출된 2000개에 대해서만 CNN을 적용하는 방식이다. 이러한 과정을 통해 기존의 방식에 비해 계산량을 줄였고, 오히려 검출률도 더 개선을 시키는 것이 가능해졌다.

하지만 이러한 기술도 저전력을 사용하는 차량 환경과 같은 임베디드 환경에 적용하기에는 적합하지 않았다. 따라서 최근 **YOLO**라는 기법이 개발되었는데, 차량 환경에서는 이 기법이 많이 이용되고 있다. 이 기법은 입력영상을 겹쳐지지 않은 그리드 형태로 쪼개어 각 그리드에 대해서만 CNN을 적용하여 R-CNN보다 훨씬 빠른 속도로 물체 검출을 수행할 수 있다. 이 방식을 통해 차량 환경에서도 실시간처리가 가능하게 되었으며, 차량이나 보행자뿐만 아니라 각종 동물이나 자전거 등 주행 상황에서 발생할 수 있는 다양한 물체에 대해 검출 및 인식이 가능하게 되었다. 이점을 해결하기 위하여 최소한의 전처리를 하도록 설계된 다 계층 퍼셉트론으로 BRF를 결합하여 기울기 감소 문제 및 시간 단축의 효과를 거둘 수 있는 **RC-CNN** 기술이 소개되기도 하였다.

최근 자율주행차를 대표하는 회사로는 테슬라(Tesla)를 꼽을 수 있을 것이다. 테슬라는 신형 전기차 ‘모델 3’에 완전 자율주행을 위한 하드웨어를 탑재한다고 발표했다. 이미 Tesla는 ‘모델 S’에 부분 자율 주행 기능을 위한 하드웨어를 탑재하고 있으며, 그간 소프트웨어 업데이트를 통해 기능을 향상시켜 왔다. Tesla는 앞으로도 온라인 소프트웨어 업데이트를 통해 사람들이 운전하는 것보다 안전하게 주행할 수 있는 자율주행자동차 개발을 목표로 하고 있다. Tesla는 모델 3 생산을 위해 생산 라인 증대 및 충전 네트워크 확대, 서비스 체제 개선, 세계 각국에서의 전시장 개선 등 다양한 준비를 진행 중이다. Tesla의 자동차가 다른 자동차와 다른 것은 자동차의 카메라와 초음파 센서 등을 장착하여 사람들의 눈처럼 빠른 속도로 물체를 감지하여 차간의 속도를 조절하거나 자동으로 차선 바꾸기, 그리고 집의 차고에 자동적으로 주차하는 ‘오토 파크 서먼(Autopark Summon)’ 기능까지 현재 구현되어 판매되고 있다는 점이다. 호우, 안개, 미세먼지 등의 기상 악화 상황 속에서도 차량 통행 유무를 쉽게 감지시킬 수 있다는 게 Tesla 측이 밝힌 완전 자율주행 오토파일럿의 특징이다.

###3. 결론 (과제에 제시된 질문을 바탕으로)

이와 같이 인공지능(AI) 기술은 여러 산업에 적용될 수 있어 최근에는 거의 모든 기업들이 인공지능 기술과 결합된 제품을 개발하기 위해 노력하고 있다고 해도 과언이 아니다. 이는 인공지능 기술을 다룰 수 있는 인재들의 필요성이 높아지고 수준 높은 인재를 영입하기 위한 기업의 경쟁도 치열해질 것이다. 필자는 개인적으로 언어분야에 관심이 많다. 현재 출시되어 있는 번역 프로그램들은 아직까지 문장에 담긴 진정한 의도까지 완벽하게 번역하는데 한계가 있다. 인공지능 기술을 활용한 번역 프로그램 등의 발전은 글로벌 비즈니스 환경에서 언어의 장벽이 허물어지는 것이므로 이는 모든 사람들이 글로벌 비즈니스를 할 수 있는 자격을 갖추게 된다는 뜻이다. 따라서, 내가 말하고 싶은 것을 언어가 통하지 않는 상대방에게 전달할 수 있는 방법이 생긴다는 것은 상상만 해도 편리하고 설레는 미래가 아닐까 생각해본다.
